```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import networkx as nx
```

```python
# Input
# A<C, D, G; B< C, D; C<H, F; D<I; E<H, I; F<H, I; G<J; J<K, L
# Dictionary, where all Value-jobs follows Key-job
job_list = {
    "A": "C, D, G",
    "B": "C, D",
    "C": "H, F",
    "D": "I",
    "E": "H, I",
    "F": "H, I",
    "G": "J",
    "J": "K, L",
}

G = nx.MultiDiGraph()  # create empty graph
for job, dependencies in job_list.items():
    for d in dependencies.split(", "):
        print(job, d)
        G.add_edge(job, d)
```

```python
# Plotting
pos = nx.spring_layout(G, iterations=10000, k=20)
node_size = 1300
label=[n for n in G.nodes()]
nodes = nx.draw_networkx_nodes(G, pos, node_size=node_size, node_color ="white", edgecolors="black")
edges = nx.draw_networkx_edges(G, pos, node_size=node_size, arrowstyle='simple',
                               arrowsize=10, edge_cmap=plt.cm.Blues, width=1, label=[n for n in G.nodes()])


# labels
# edeges
# nx.draw_networkx_edge_labels(G,pos=pos, font_size=20, edge_labels={('A','B'): "A"})
# nodes
nx.draw_networkx_labels(G, pos, font_size=20, font_family='sans-serif')
# ax.set_axis_off()
plt.show()
```

```python
pix = np.array(img)
left = 14
top = 14
right = 0
bot = 0
for ir, row in enumerate(pix):
    for ip, p in enumerate(row):
        # print(p)
        if p.any():
            # print(p)
            # print(ip, left)
            if sum(pix[ir][ip])/3 >= 180:
                pix[ir][ip] = [255,255,255]
            else:
                pix[ir][ip] = [0,0,0]
            left = min(ip, left)
            right = max(ip, left)
            top = min(ir, top)
            bot = max(ir, bot)
print(left)
print(right)
print(top)
print(bot)
im1 = Image.fromarray(pix).crop((left, top, right, bot))
# s = im1.thumbnail((15,15), Image.ANTIALIAS)
s = im1.resize((15,15))
s.save("A15.png")
```
